```python
parkinsons = parkinsons.drop(columns=['name'])#dropping rows with missing values
```

```python
# Handle missing values
x = parkinsons.dropna()# remove missing values
```

```python
# Separate Features and Target
x = parkinsons.drop(['status'],axis=1)
y = parkinsons['status']
```

```python
#Encoding categorical variables(if neccessary)
x =pd.get_dummies(x, drop_first=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

```

```python
# Train-test split
x_train_p, x_test_p, y_train_p, y_test_p = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train_p, y_train_p)
y_pred_rf = rf_model.predict(x_test_p)
```

```python
from xgboost import XGBClassifier

# Remove the 'use_label_encoder' parameter, as it's not needed in the latest versions of XGBoost
xgbparkinsons_model = XGBClassifier(eval_metric='logloss')
xgbparkinsons_model.fit(x_train_p, y_train_p)

# Make predictions and calculate accuracy for XGBoost
y_pred_xgb = xgbparkinsons_model.predict(x_test_p)

# Accuracy for XGBoost
xgb_accuracy = accuracy_score(y_test_p, y_pred_xgb)
print(f"XGBoost Accuracy: {xgb_accuracy:.2f}")

# Classification Report and Confusion Matrix for XGBoost
print("XGBoost Classification Report:\n", classification_report(y_test_p, y_pred_xgb))
print("XGBoost Confusion Matrix:\n", confusion_matrix(y_test_p, y_pred_xgb))
```

```python
# Random Forest Accuracy
rf_accuracy = accuracy_score(y_test_p, y_pred_rf)
print(f"Random Forest Accuracy: {rf_accuracy:.2f}")
```

```python
# Accuracy for XGBoost
xgb_accuracy = accuracy_score(y_test_p, y_pred_xgb)
print(f"XGBoost Accuracy: {xgb_accuracy:.2f}")
```

```python
# Classification Report
print("Random Forest Classification Report:\n", classification_report(y_test_p, y_pred_rf))
print("XGBoost Classification Report:\n", classification_report(y_test_p, y_pred_xgb))
```

```python
# Confusion Matrix
print("Random Forest Confusion Matrix:\n", confusion_matrix(y_test_p, y_pred_rf))
print("XGBoost Confusion Matrix:\n", confusion_matrix(y_test_p, y_pred_xgb))
```
